```python
from TTS.tts.configs.shared_configs import BaseDatasetConfig
from TTS.tts.datasets import load_tts_samples
import os
global path
path = "/content/drive/MyDrive/Implementation/TTS/VITS_Output"
```

```python
import pandas as pd
from arabic_pronounce import phonetise

def formatter(root_path, manifest_file, **kwargs):  # pylint: disable=unused-argument
    items=[]
    speaker_name= "Arabic_man"
    with open(manifest_file, 'r', encoding='utf-16') as file:
      data = file.readlines()
      data = [line.strip().split('|') for line in data]
      data = pd.DataFrame(data, columns=['filename', 'text'])
      for _, row in data.iterrows():
        row['filename'] += '.wav'
        wav_file=root_path+row['filename']
        entire_text = ""
        for text in row['text'].split():
          entire_text+="".join(phonetise(text)[0])+" + " ## New Change
        items.append({"text":entire_text[:-3], "audio_file":wav_file, "speaker_name":speaker_name, "root_path": root_path})
    return items
```

```python
train_dataset_config = BaseDatasetConfig(
    formatter="", meta_file_train="/content/ClarTTS_corpus/training.txt", language="", path="/content/ClarTTS_corpus/wav/train/",meta_file_val='/content/ClarTTS_corpus/wav/val')

train_samples,_= load_tts_samples(datasets=train_dataset_config,formatter=formatter,eval_split=False)
```
